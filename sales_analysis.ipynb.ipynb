```python
import pandas as pd

# Load CSV from the extracted folder
data = pd.read_csv("C:/Users/diksh/Documents/archive/Sample - Superstore.csv", encoding='latin1')

# Display first 5 rows
data.head()
```

```python
data.info()
data.describe()
```

```python
data.isnull().sum()
```

```python
# Unique values in key columns
print("Regions:", data['Region'].unique())
print("Categories:", data['Category'].unique())
print("Sub-Categories:", data['Sub-Category'].unique())
```

```python
region_summary = data.groupby('Region')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
region_summary
```

```python
top_products = data.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
category_profit = data.groupby('Category')['Profit'].sum()
category_profit
```

```python
import matplotlib.pyplot as plt
region_summary = data.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
region_summary.plot(kind='bar', figsize=(8,5))
plt.title('Total Sales and Profit by Region')
plt.xlabel('Region')
plt.ylabel('Amount')
plt.show()
```

```python
category_sales = data.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_sales.plot(kind='bar', color='skyblue', figsize=(7,4))
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
subcat_profit = data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
subcat_profit.head(10).plot(kind='bar', color='orange', figsize=(10,5))
plt.title('Top 10 Sub-Categories by Profit')
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.show()
```

```python
# Convert order date to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Group by month
monthly_sales = data.groupby(data['Order Date'].dt.to_period('M'))['Sales'].sum()

monthly_sales.plot(figsize=(10,5))
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
